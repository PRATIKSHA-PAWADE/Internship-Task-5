```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
at=pd.read_csv("Attrition.csv")
at.head()
```

```python
at.columns
```

```python
at.shape
```

```python
at["Attrition"].unique()
```

```python
at["BusinessTravel"].unique()
```

```python
sns.catplot(x='BusinessTravel' , data=at , kind='count', hue='Attrition')
plt.ylabel("Count of Employee")
plt.xlabel("Categories of business travel")
plt.title("Relation between Attrion and BusinessTravel" , color ="limegreen")
```

```python
at.Attrition.replace({"Yes":1, "No":0},inplace=True)
```

```python
df=pd.DataFrame(at.groupby('BusinessTravel').Attrition.sum())
df= df.sort_values('Attrition')
```

```python
plt.bar(df.index , df.Attrition)
plt.title("Relation between Attrion and BusinessTravel" , color ="magenta")
```

```python
plt.pie(df.Attrition , labels = df.index , autopct ="%1.02f%%")
plt.title("Relation between Attrion and BusinessTravel" , color ="darkblue")
plt.show()
```
